### Polar DE

```python
def model(r, t):
    # Define the differential equation
    r = 15 * t  # Update r to be a function of time
    drdt = 15
    dsdt = 30
    dthetadt = np.sqrt((dsdt**2 - drdt**2)/(r**2))
    
    return dthetadt

# Set the initial value of r
r0 = 3

# Set the time points at which the solution should be calculated
t = np.linspace(0, 1, 10)

# Calculate the solution to the differential equation
r = odeint(model, r0, t+0.000000001)  # Update odeint() call to pass time as an additional argument

# Plot the solution
import matplotlib.pyplot as plt

plt.plot(t, r)
plt.xlabel('Time')
plt.ylabel('r')
plt.show()
```

```python
# in terms of theta not r

def model(theta, t):
    # Define the differential equation
    r = 3  # Update r to be a function of theta
    drdt = 15
    dsdt = 30
    dthetadt = np.sqrt((dsdt**2 - drdt**2)/(r**2))
    
    return dthetadt

# Set the initial value of theta
theta0 = 0

# Set the time points at which the solution should be calculated
t = np.linspace(0, 10, 100)

# Calculate the solution to the differential equation
theta = odeint(model, theta0, t)  # Update odeint() call to pass time as an additional argument

# Plot the solution
import matplotlib.pyplot as plt

plt.plot(t, theta)
plt.xlabel('Time')
plt.ylabel('theta')
plt.show()
```

### Logarithmic Spiral (spiral of Archimedes)

Destroyer equation: $r=3e^\frac{\theta}{\sqrt{3}}$

```python
### plot logarithmic spiral ###

# define the function

def r(theta): return 3*np.exp(theta/np.sqrt(3))

# define the range of theta

theta = np.linspace(0, 4*math.pi, 1000)

# plot the curve

plt.plot(r(theta)*np.cos(theta), r(theta)*np.sin(theta), 'r')
plt.gca().set_aspect('equal', adjustable='box') # center plot and set equal aspect ratio
# plt.axhline(y=0, color='k') # axis lines
# plt.axvline(x=0, color='k')
plt.grid(color='b', linestyle='-', linewidth=0.1)
plt.show()
```

```python
### zooming in on the intercept

# define the function

def r(theta): return 3*np.exp(theta/np.sqrt(3))

# define the range of theta

theta = np.linspace(0, math.pi, 1000)

# plot the curve

plt.plot(r(theta)*np.cos(theta), r(theta)*np.sin(theta), 'r')
plt.gca().set_aspect('equal', adjustable='box') # center plot and set equal aspect ratio

# plot straight line with theta2 = 2pi/3

theta2 = 2*math.pi/3
r2 = np.linspace(0, 10, 10)
plt.plot(r2*np.cos(theta2), r2*np.sin(theta2), 'b')

plt.grid(color='b', linestyle='-', linewidth=0.1)
plt.show()
```

Demonstrate that this solution eq makes sense because
1) logarithmic curves have special property that the tangent line is always perpendicular to the curve, and 
2)  radius of curvature increases linearly with the arc length.

```python
# https://en.wikipedia.org/wiki/Logarithmic_spiral#/media/File:Logspiral.gif

theta = np.linspace(0, 2 * np.pi, 100)

# Define the radius of the circle and the logarithmic spiral
r_circle = 3
a = 3  # initial size of the spiral
b = 0.2  # rate at which the spiral grows
r_spiral = a * np.exp(b * theta)  # The radius of the spiral is a function of theta

# Convert the polar coordinates to cartesian coordinates
x_circle = r_circle * np.cos(theta)
y_circle = r_circle * np.sin(theta)
x_spiral = r_spiral * np.cos(theta)
y_spiral = r_spiral * np.sin(theta)

# Plot the circle and the spiral
plt.plot(x_circle, y_circle, label='Circle')
plt.plot(x_spiral, y_spiral, label='Spiral')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
